```python
results_dict = {}

for ticker in nifty_500_tickers:
    df = yf.download(ticker, period="251d")
    if df.empty:
        continue
    df = df.reset_index()
    df.columns = ['date', 'close', 'high', 'low', 'open', 'volume']
    df = df[df['volume'] != 0]
    df['5EMA'] = df['close'].ewm(span=5, adjust=False).mean()
    df['9EMA'] = df['close'].ewm(span=9, adjust=False).mean()
    df['20EMA'] = df['close'].ewm(span=20, adjust=False).mean()
    df['21EMA'] = df['close'].ewm(span=21, adjust=False).mean()
    df['50SMA'] = df['close'].rolling(window=50).mean()
    if not df.empty:
        record = df.tail(1).to_dict('records')[0]
        record = {k: round(v, 2) if isinstance(v, (int, float, np.float64, np.int64)) else v for k, v in record.items()}
        record = {'ticker': ticker, **record}
        results_dict[ticker] = record

results_df = pd.DataFrame.from_dict(results_dict, orient='index')
results_df.reset_index(drop=True, inplace=True)
cols = ['ticker'] + [col for col in results_df.columns if col != 'ticker']
results_df = results_df[cols]
results_df.head()
```

```python
final_df = results_df[
    (results_df['9EMA'] > results_df['21EMA']) &
    (results_df['5EMA'] > results_df['20EMA']) &
    (results_df['20EMA'] > results_df['50SMA']) &
    (results_df['close'] > results_df['50SMA'])
]
```

```python

final_df['trend_distance_1'] = ((final_df['20EMA'] - final_df['50SMA']) / final_df['50SMA']) * 100
final_df['trend_distance_2'] = ((final_df['9EMA'] - final_df['21EMA']) / final_df['21EMA']) * 100
final_df['trend_distance_3'] = ((final_df['5EMA'] - final_df['20EMA']) / final_df['20EMA']) * 100
final_df['final_trend_distance'] = (
    0.5 * final_df['trend_distance_1'] +
    0.3 * final_df['trend_distance_2'] +
    0.2 * final_df['trend_distance_3']
)
final_df = final_df.sort_values(by='final_trend_distance', ascending=False)

```
